### Data Cleaning
#### Objective: To submit a cleaned and standardized list of customers to the company care team who will find it easier making the calls to the customers.

```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
df
```

#### Dropping duplicates

```python
df = df.drop_duplicates()
df
```

#### Dropping unnecessary columns

```python
df = df.drop(['Not_Useful_Column'], axis=1) # or df.drop(columns='Not_Useful_Column')
df
```

#### Cleaning specific column Data

```python
#Using strip method to remove spaces from both left and right side of the string
#first method: removing individually
#df['Last_Name'] = df['Last_Name'].str.lstrip('...')
#df['Last_Name'] = df['Last_Name'].str.lstrip('/')
#df['Last_Name'] = df['Last_Name'].str.rstrip('_')

#Second method: removing Collectively
df['Last_Name'] = df['Last_Name'].str.strip('123./')
df
```

#### Cleaning and Standardizing the phone numbers

##### Step 1: Getting rid of non-numeric values 
#####  using regular expressions to get rid of non-numeric values meaning: replacing evrything that is not abc/ABC or an integer value with an empty str. 

```python
#converting column to string before performing regex
df = df.astype({'Phone_Number': 'string'}) 
```

```python
df.info()
```

```python
# removing non-str and non-int values and replacing with an empty string.
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^0-9]','',regex=True) 
df
```

##### Step 2: Reformat the numbers to the standardized approach.

```python
df['Phone_Number'].apply( lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10] )
```
