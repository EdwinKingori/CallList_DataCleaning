### Data Cleaning
#### Objective: To submit a cleaned and standardized list of customers to the company care team who will find it easier making the calls to the customers.

```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
df
```

### Dropping duplicates

```python
df = df.drop_duplicates()
df
```

### Dropping unnecessary columns

```python
df = df.drop(['Not_Useful_Column'], axis=1) # or df.drop(columns='Not_Useful_Column')
df
```

#### Cleaning specific column Data

```python
#Using strip method to remove spaces from both left and right side of the string
#first method: removing individually
#df['Last_Name'] = df['Last_Name'].str.lstrip('...')
#df['Last_Name'] = df['Last_Name'].str.lstrip('/')
#df['Last_Name'] = df['Last_Name'].str.rstrip('_')

#Second method: removing Collectively
df['Last_Name'] = df['Last_Name'].str.strip('123./')
df
```

### 1. Cleaning and Standardizing the Phone_Number Column

##### Step 1: Getting rid of non-numeric values 
#####  using regular expressions to get rid of non-numeric values meaning: replacing evrything that is not abc/ABC or an integer value with an empty str. 

```python
#converting column to string before performing regex
df = df.astype({'Phone_Number': 'string'}) 
```

```python
df.info()
```

```python
# removing non-str and non-int values and replacing with an empty string.
df['Phone_Number'] = df['Phone_Number'].str.replace(r'[^0-9]','',regex=True) 
df
```

```python
df.info()
```

##### Step 2: Reformat the numbers to the standardized approach.

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df['Phone_Number']
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x:x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

##### Step3 : Removing both the  'NA' and '--' values in the Phone_Number column

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('<NA->-','')

df['Phone_Number'] = df['Phone_Number'].str.replace('--','')

df.Phone_Number
```

### Splitting the Address column into stree, state & zip_code string indices

```python
df.head()
```

```python
df[['Street', 'State', 'Zip_Code']] = df['Address'].str.split(',', expand=True) 
df
```

### Cleaning and standardizing the strings within the 'Paying Customer & Do_Not_Contact' columns.

```python
#Changing the 'Y' string into 'Yes'
df['Paying Customer'].str.replace('Y', 'Yes')
```

```python
#Now Changing the 'Yeses' & 'Yes' strings into 'Y'
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y') 

#Changing the 'No' string to 'N'
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N') 
df
```

```python
# Changing the 'Do_Not_Contact' column's strings
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y') 

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N') 
df
```

```python
#df.replace('N/a','')
#df.replace('NaN','')

# Removing and replacing the "N/a" and "NaN"  with a blank space 
df = df.fillna('')
df
```

```python
# removing 'N/a'in row 18
df = df.replace('N/a','')
df
```

#### The call center demands only calls that can be contacted. Thus, excluding the "yes" responses in the do_not_contact column and removing the empty cells in the phone number column. 

```python
# Excluding the "yes" responses in the do_not_contact column
for i in df.index:
    if df.loc[i, 'Do_Not_Contact'] == 'Y':
        df.drop(i, inplace=True)
df
```

```python
# Removing the empty cells in the phone number column.
for i in df.index:
    if df.loc[i, 'Phone_Number'] == '':
        df.drop(i, inplace=True)
df

#Another approach of dropping null values
#df = df.dropna(subset='Phone_Number', inplace=True)
```

##### Can also use this epression: df[(df["Do_Not_Contact"]  == 'No') & (df["Phone_Number"] != '')] , to meet the conditions above.

```python
#lets replace the blanks in the 'Do_Not_Contact' column with 'N'
df['Do_Not_Contact'] = df['Do_Not_Contact'].replace('', 'N')
df
```

```python
# Resetting the index before submitting the doc to the contact center
df = df.reset_index(drop=True)
df
```
